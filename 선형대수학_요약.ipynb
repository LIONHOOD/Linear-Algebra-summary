# 선형대수학 요약

- **기저(basis)** : $n$차원 벡터공간에서 일차결합(linear combination)을 통해 모든 벡터를 표현할 수 있는 $n$개의 **일차독립(linearly independent)**인 벡터 집합
 - **정규직교기저(orthonormal basis)** : 기저에 속한 모든 벡터가 단위벡터이고, 모든 서로 다른 두 벡터의 **내적(inner product)**이 0인 기저
 - **그람-슈미트 정규직교화(Gram-Schmidt orthonormalization)** : 일차독립인 벡터 집합을 정규직교기저로 변환하는 방법
   - 주어진 벡터에서 기준 벡터로의 정사영을 빼서 기준 벡터에 직교하는 벡터를 추출

- **고유벡터(eigenvector)** : **선형변환(linear transformation)**된 벡터와 상수(**고윳값(eigenvalue)**)배가 동일한 0이 아닌 벡터
>$Av=λv$
 - 선형변환은 그에 대응되는 행렬을 곱하는 것과 동치
- **직교행렬(orthogonal matrix)** : 행벡터와 열벡터가 유클리드 공간의 정규직교기저를 이루는 행렬  
$⟺$ 전치행렬과 역행렬이 동일한 행렬
>$A^T=A^{-1}$
- **대각행렬(diagonal matrix)** : 주대각성분을 제외한 모든 성분이 0인 정사각행렬
- **대칭행렬(symmetric matrix)** : 전치행렬과 동일한 행렬
>$A^T=A$
 - 주어진 행렬 $M$에 대하여, $M^TM$은 대칭행렬이고 모든 고윳값이 음이 아닌 positive semi-definite(양의 준정부호) 행렬
- **직교대각화가능(orthogonally diagonalizable)** : $P^{-1}AP(=P^TAP)=D$ (단, $D$는 대각행렬)를 만족하는 직교행렬 $P$가 존재할 때 행렬 $A$는 직교대각화가능  
$⟺$ $A$가 대칭행렬  
$⟺$ $A$의 고유벡터로 구성된 정규직교기저가 존재(**eigendecomposition**)
 - $P$는 열벡터가 $A$의 고유벡터로 구성된 행렬, $D$는 주대각성분이 순서대로 대응되는 고윳값으로 구성된 대각행렬
 - eigendecomposition을 직사각행렬에 적용 가능하도록 일반화한 개념이 SVD(singular value decomposition); 고유벡터(고윳값) $→$ 특이벡터(특잇값)
- **레일리 몫(Rayleigh quotient)** : $A$가 대칭행렬이고 $λ_{min}(λ_{max})$가 $A$의 최소(최대) 고윳값일 때, 0이 아닌 임의의 벡터 x에 대하여 아래 부등식이 항상 성립
>$λ_{min}\leqslant R(A,x)=\frac{x^TAx}{x^Tx}\leqslantλ_{max}$
 - 좌측 등호 : $x=v_{min}$($λ_{min}$에 대응되는 고유벡터)
 - 우측 등호 : $x=v_{max}$($λ_{max}$에 대응되는 고유벡터)
 - 주어진 행렬 $M$에 대하여, $M^TM=A$는 대칭행렬이므로 $\sqrt{R(A,x)}=\frac{\|Mx\|}{\|x\|}$의 최대를 찾는 문제에 적용 가능
 - <라그랑주 승수법에 의한 증명>  
 1. $R(A,cx) = \frac {(cx)^{T} A cx} {(cx)^{T} cx} = \frac {c^{T} c} {c^{T} c} \frac {x^{T} A x} {x^{T} x} = R(A,x)$이므로 $x$를 단위벡터로 가정할 수 있고 제약 조건은 $\|x\|^2=x^Tx=1$
 2. 함수 $R(A,x)=x^TAx$ 와 라그랑주 승수 $\lambda$에 대해 $L=x^{T}Ax-\lambda(x^{T}x-1)$라 두면, $\frac{dL}{dx}=0⇒Ax=λx⇒R(A,x)=λ$
 3. 따라서, 레일리 몫의 임계점은 $x$가 $A$의 고유벡터인 경우로 임계값은 대응되는 고윳값

**Notation 대응 참고(복소수 확장)**
- 전치(transpose) $T$ : 켤레전치(conjugate transpose) $*$
- 직교 : 유니터리(unitary)
- 대칭행렬 : 에르미트행렬(Hermitian matrix)
